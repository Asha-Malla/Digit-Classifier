```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
len(X_test)
```

```python
len(y_test)
```

```python
X_train[0].shape
```

```python
X_train[0]
```

```python
plt.matshow(X_train[0])
```

```python
y_train[0]
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_tr=X_train.reshape(len(X_train),28*28)
X_ts=X_test.reshape(len(X_test),28*28)
```

```python
X_ts.shape
```

```python
X_tr[0]
```

```python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation="sigmoid")
])
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=['accuracy'])
model.fit(X_tr,y_train,epochs=5)
```

```python
model.evaluate(X_ts,y_test)
```

```python
plt.matshow(X_test[0])
```

```python
res=model.predict(X_ts)
res[0]
```

```python
np.argmax(res[0])
```

```python
y_pred_labels=[np.argmax(i) for i in res]
```

```python
print(y_pred_labels)
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred_labels)
cm
```

```python
!pip install seaborn
```

```python
import seaborn as sb
plt.figure(figsize=(10,7))
sb.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('Truth')
```
